문제 요약

과학자가 발표한 논문의 인용횟수를 담은 배열 citations가 주어집니다. 

발표한 논문 n편 중, h번 이상 인용된 논문이 h편 이상이고, 나머지 논문이 h번 이하 인용되었다는 조건을 만족하는 h의 최대값을 출력해야 합니다.

```python
! # 끝 시간 체크
```

### 풀이

```python
from bisect import bisect_left

def solution(citations):
    citations.sort()
    result=0
    for h in range(len(citations)+1):
        h_papers = len(citations[bisect_left(citations, h):])
        if h_papers >= h:
            result = h
    return result
```
