문제 요약

    숫자 배열이 주어질 때, 순서를 재배치하여 만들 수 있는 가장 큰 수를 문자열로 리턴해야 합니다.

```python
! # 끝 시간 체크
```

### 풀이

```python
def solution(numbers):    
    s = [*map(str,numbers)]
    a=sorted(s,key=lambda x: x*4,reverse=1)
    return str(int("".join(a)))
```

### 좀 더 짧은 풀이

```python
def solution(numbers):    
    return str(int(''.join(sorted(map(str,numbers), key=lambda x: x*3)[::-1])))
```
