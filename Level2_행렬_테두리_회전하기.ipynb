### 주석 달린 풀이

```python
def solution(rows, columns, queries):
    answer = []
    arr=[];c=0
    for _ in range(rows): # arr 생성
        arr.append([(c:=c+1) for _ in range(columns)])

    for q in queries:
        r1,c1,r2,c2 = [i-1 for i in q] # zero index로 변환
        m = first = arr[r1][c1] # r1, c1 (초기값) 저장 및 최소값 초기화

        for r in range(r1,r2): # 1. 위로 이동
            k = arr[r][c1] = arr[r+1][c1]
            m = min(m,k)

        for c in range(c1,c2): # 2. 왼쪽으로 이동
            k = arr[r2][c] = arr[r2][c+1]
            m = min(m,k)

        for r in range(r2,r1,-1): # 3. 아래로 이동
            k = arr[r][c2] = arr[r-1][c2]
            m = min(m,k)

        for c in range(c2,c1,-1): # 4. 오른쪽으로 이동
            k = arr[r1][c] = arr[r1][c-1]
            m = min(m,k)

        arr[r1][c1+1] = tmp # *빨간 동그라미 부분 (초기값으로 채우기)
        answer.append(m) # 최소값 answer에 추가
    return answer
```
