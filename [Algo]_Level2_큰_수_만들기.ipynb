### 내 풀이

```python
def solution(number, k):
    stack=[]
    K=k
    for n in number:
        while stack and K and stack[-1] < n:
            stack.pop()
            K-=1
        stack.append(n)
    if K:
        return number[:-k]
    else:
        return "".join(stack)
```

### 다른 사람 풀이

```python
def solution(number, k):
    st = []
    for n in number:
        while st and k > 0 and st[-1] < n:
            st.pop()
            k -= 1
        st.append(n)
    return ''.join(st[:len(number)-k])
```
